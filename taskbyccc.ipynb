```python
# importing python libraries for EDA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/Crimes_In_India_2001_2012.csv')

```

```python
df.head()
```

```python
dnew= df[df['DISTRICT'].str.contains('TOTAL')==False]
```

```python
fig,axes=plt.subplots(figsize=(15,5)) 
sns.boxplot(x="YEAR", y="TOTAL IPC CRIMES", data=dnew, palette='rainbow')       # Total ipc crimes per year ( including all states together)   
```

```python
print(dnew['STATE/UT'].unique())
```

```python
dnew.tail(1)
```

 Crimes in each state representation

```python
#  crimes in each state representation


g = dnew.groupby('STATE/UT')['TOTAL IPC CRIMES'].sum()
sns.set()
plt.figure(figsize=(24, 6))
g.plot(kind = 'bar', color = 'blue', alpha = 0.90, width = 0.50)
```

```python
swdf = dnew['STATE/UT'].value_counts().sort_index()
swdf
```

```python
dnew.sort_values(by = 'TOTAL IPC CRIMES', ascending = False, inplace = True)
dnew.head()
```

```python
ds = dnew[['STATE/UT','TOTAL IPC CRIMES']].copy()
ds.head()
```

 Depicting 10 states with the highest number of IPC crimes.

```python
#  Depicting 10 states with the highest number of IPC crimes.




dss = ds.groupby('STATE/UT', sort = False).sum()
dsss = dss.sort_values(by = 'TOTAL IPC CRIMES', ascending = False)
dsss.head(10).plot(kind = 'bar', color = 'blue', figsize = (15,8), title = "Most number of IPC Crimes")
```

 District wise distribution of crimes in  Arunachal Pradesh, Uttar Pradesh and West Bengal for the year 2001 and 2004.


```python
df_Andra = dnew[dnew['STATE/UT'].str.contains('ANDHRA PRADESH')==True]
df_Andra_2001 = df_Andra[df_Andra['YEAR'] <= 2001]
```

```python
df_Andra_2001_plot = df_Andra_2001.groupby('DISTRICT')['TOTAL IPC CRIMES'].sum()
sns.set()
plt.figure(figsize=(20, 10))
df_Andra_2001_plot.plot(kind = 'bar', color = 'blue', alpha = 0.90, width = 0.70)
```

```python
df_Andra_2001_plot = df_Andra_2001.groupby('DISTRICT')['TOTAL IPC CRIMES'].sum()
sns.set()
plt.figure(figsize=(15, 15))
df_Andra_2001_plot.plot(kind = 'pie')
```

```python
df_Andra_2004 = df_Andra[(df_Andra['YEAR'] >= 2004) & (df_Andra['YEAR'] < 2005)]
df_Andra_2004_plot = df_Andra_2004.groupby('DISTRICT')['TOTAL IPC CRIMES'].sum()
sns.set()
plt.figure(figsize=(20, 10))
df_Andra_2004_plot.plot(kind = 'bar', color = 'blue', alpha = 0.90, width = 0.70)
```

```python
df_Andra_2004_plot = df_Andra_2004.groupby('DISTRICT')['TOTAL IPC CRIMES'].sum()
sns.set()
plt.figure(figsize=(15, 15))
df_Andra_2004_plot.plot(kind = 'pie')
```

  crimes committed against children and women.

```python
dn = dnew[['STATE/UT', 'YEAR', 'RAPE', 'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS', 'DOWRY DEATHS', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY', 'INSULT TO MODESTY OF WOMEN', 'CRUELTY BY HUSBAND OR HIS RELATIVES', 'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES']].copy()
dn['Crime Against Women'] = dn['RAPE']+dn['KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS']+dn['DOWRY DEATHS']+dn['ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY']+dn['INSULT TO MODESTY OF WOMEN']+dn['CRUELTY BY HUSBAND OR HIS RELATIVES']+dn['IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES']
dn.head()
```

```python


dns = dn.groupby('STATE/UT', sort = False)['Crime Against Women'].sum()
sns.set()
plt.figure(figsize=(20, 10))
dns.plot(kind = 'bar', color = 'blue', alpha = 0.90, width = 0.70)
```

It can be easily seen that andhra pradesh has the highest no. of crimes against childrens and womens.
